# Data Wrangling


- Data wrangling also known as data munging, is the process of transforming and mapping of data from its atomic data form into a format whereby it makes the extracted information more apppropriate and valuable for a variety of purposes such as analytics.

## Wrangling Data by:

- Merging data frames
- Filtering data frames
- Manipulating rows and columns

## Importing essential python libraries 


```python
import pandas as pd
```

## Creating a data frame from importing the personality score dataset

```python
pscore = pd.read_csv('/home/rulani/Downloads/Data-wrangling-master/personality_scores.csv',delimiter=";")
pscore.dropna(axis='columns').head()
```

## Examining data for :

- Uniqueness of ID 
- Availability of NaN entires

```python
# dropping ALL duplicte values 
dropped_pscore = pscore.drop_duplicates(subset ="ID", 
                     keep = False, inplace = False) 
```

```python
#Checking the lengths of the datasets using assert

def same():
    assert len(pscore) == len(dropped_pscore)
    return True
       
same()
```

```python
print('There are no duplicated IDs in this data frame since '
    '\nThe length of dropped duplicate data frame is,' ,len(dropped_pscore),
    '\nand that of the original data frame is', len(pscore), '.')
```

```python
# making new data frame with dropped NA values 
new_dropped_pscore = dropped_pscore.dropna(axis = 1, how ='any', inplace = False)

new_dropped_pscore.head()
```

## Creating a function for the total score of each of the personality test subscales

```python
def assigning(row,Category_Num):
    count = 0
    asd = 0
    for column in new_dropped_pscore.columns:
        tuplee = row[column]
        if type(tuplee) == str:
            catergory = tuplee[1]
            result = int(tuplee[4])
            if catergory == Category_Num:
                count += result
    return count
```

## Creating a column for the Big-Five Factor Markers

```python
new_dropped_pscore['Extraversion'] = dropped_pscore.apply(lambda row : assigning(row,'1'), axis = 1)
new_dropped_pscore['Agreeableness'] = dropped_pscore.apply(lambda row : assigning(row,'2'), axis = 1)
new_dropped_pscore['Conscientiousness'] = dropped_pscore.apply(lambda row : assigning(row,'3'), axis = 1)
new_dropped_pscore['Emotional Stability'] = dropped_pscore.apply(lambda row : assigning(row,'4'), axis = 1)
new_dropped_pscore['Openness to experiences'] = dropped_pscore.apply(lambda row : assigning(row,'5'), axis = 1)
```

```python
#Displaying the data frame incoporated with the big five factor Markers

new_dropped_pscore[['ID','Extraversion','Agreeableness',
                    'Conscientiousness','Emotional Stability',
                    'Openness to experiences']].tail()
```
