#### Importing

```python
import pandas as pd
#import unittest 
from pandas.util.testing import assert_frame_equal
```

#### Reading in data

```python
pscore = pd.read_csv('personality_scores.csv',delimiter=";")
pscore.dropna(axis='columns').head(2)
```

#### Examining datasets for duplicates

```python
# sorting by first name 
pscore.sort_values("ID", inplace = True)
pscore

# dropping ALL duplicte values 
dropped_pscore = pscore.drop_duplicates(subset ="ID", 
                     keep = False, inplace = False) 
len(dropped_pscore)
```

#### Checking the lengths of the datasets using assert

```python
#assertEqual(pscore, dropped_pscore)

def same(length):
    assert len(pscore) == len(dropped_pscore)
    return True
       
same('')
```

#### Creating columns containing the total score of each of the personality test subscales

```python
def assigning(row,Category_Num):
    count = 0
    asd = 0
    for column in dropped_pscore.columns:
        tuplee = row[column]
        if type(tuplee) == str:
            catergory = tuplee[1]
            result = int(tuplee[4])
            if catergory == Category_Num:
                count += result
    return count
```

```python
dropped_pscore['Extraversion'] = dropped_pscore.apply(lambda row : assigning(row,'1'), axis = 1)
dropped_pscore['Agreeableness'] = dropped_pscore.apply(lambda row : assigning(row,'2'), axis = 1)
dropped_pscore['Conscientiousness'] = dropped_pscore.apply(lambda row : assigning(row,'3'), axis = 1)
dropped_pscore['Emotional Stability'] = dropped_pscore.apply(lambda row : assigning(row,'4'), axis = 1)
dropped_pscore['Openness to experiences'] = dropped_pscore.apply(lambda row : assigning(row,'5'), axis = 1)
```
